```python
%pip install pyspark

from pyspark.sql import SparkSession
import pyspark.sql.functions as F

spark = SparkSession.builder.appName("RDD Native").master("spark://localhost:7077").getOrCreate()
print(spark)
```

```python
rdd_data = [
    ("John", 28),
    ("Jane", 32),
    ("Doe", 45),
    ("Ritik", 28)
]

rdd = spark.sparkContext.parallelize(rdd_data)
rdd.getNumPartitions()
```

```python
rdd = rdd.coalesce(4)
rdd.getNumPartitions()
```

```python

rdd.collect()
```

```python
rdd = rdd.map(lambda x: (x[0], x[1]  + 5))
rdd.collect()
```

```python
rdd.groupBy(lambda x: x[0]).collect()
```

```python
rdd2_data = [
    ("John", "USA"),
    ("Jane", "UK"),
    ("Doe", "India"),
    ("Ritik", "India")
]

rdd2 = spark.sparkContext.parallelize(rdd2_data)
rdd2 = rdd2.coalesce(4)
rdd2.collect()
```
