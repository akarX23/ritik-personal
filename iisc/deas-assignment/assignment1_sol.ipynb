#### Question 1.2: Content cleanup **(15 points)**

Clean the the `content` column by performing the folloing tasks:
1. Convert all text of content column to lowercase.
2. Remove standard HTML tags.
3. Replace the following escape sequences present as literal in the content column - `\n, \t, \\, \', \", \r, \b, \f, \a` with whitespace.
4. Replace consecutive whitespaces with a single whitespace. Trim leading and trailing whitespaces.
5. Remove stopwords for content in english language (stopwords are given in `stopword_df`).

**Input:** You are given the dataframe from the output of *Q1.1* having following schema:

`input_df`

*  `uri: string`
*  `lang: string`
*  `date: string`
*  `content: string`
*  `crawl_month: string`

`stopword_df`
*  `stopwords`

**Output:** Return an output dataframe that retains some of the input dataframe columns, but with the `content` cleaned up as per the tasks given above. The output DF should have the columns:

*  `uri: string`
*  `lang: string`
*  `crawl_month: string`
*  `content: string`

```python
#######################################
###!@1.2 START ANSWER SET 1.2

### Q1.2 ###################################################

def question_1_2(input_df, stopword_df):

  ## start your edits here =================

  escape_seq = ['\\n', '\\t', '\\r', '\\b', '\\f', '\\a', '\\\\', '\\"', "\\'"]
  escape_seq_reg_expr = create_regex_from_list(escape_seq)
  stopwords_list = stopword_df.select("stopwords").rdd.flatMap(lambda x: x).collect()
  punctuations = [".", ",", "!", "?", ";", ":", "(", ")", "{", "}", "[", "]", "'", '"', "<", ">", "/", " "]
  punctuation_pattern = "([" + "".join(re.escape(c) for c in punctuations) + "])"

  sc = SparkContext.getOrCreate()
  punctuation_pattern = sc.broadcast(punctuation_pattern)
  stopwords_list = sc.broadcast(stopwords_list)

  input_df = input_df.drop("date").withColumn("content", F.lower(F.col("content"))) #lowercase
  input_df = input_df.withColumn("content", F.regexp_replace(F.col("content"), r'<.*?>', " ")) #HTML Tags
  input_df = input_df.withColumn("content", F.regexp_replace(F.col("content"), escape_seq_reg_expr, " ")) # Escape Seqs
  input_df = input_df.withColumn("content", F.trim(F.regexp_replace(F.col("content"), r'\s+', ' '))) # Whitespaces
  en_input_df = input_df.where(F.col("lang") == "en")
  en_input_df = en_input_df.withColumn("content", remove_stopwords(F.col("content"), F.lit(stopwords_list.value), F.lit(punctuation_pattern.value))) # Stopwords
  input_df = input_df.where(F.col("lang") != "en").union(en_input_df)

  output_df = input_df.select("uri", "lang", "crawl_month", "content")

  ## end your edits here =================

  return output_df

###!@1.2 END ANSWER SET 1.2
```

#### Question 2.2: Calculate the frequency of documents for each keyword **(15 points)**
You are given with two dataframes: `input_df (uri, lang, crawl_month, content)` and an additional dataframe `keyword_df (keyword)` having topics of interest. You have to do the following:
1. Count the number of documents containing the given keywords, for each keyword in each month.
2. Compute the normalized frequency of documents containing keyword per month using the formula:<br>
 $normalized\_doc\_count = \frac{Number\_of\_docs\_with\_given\_keyword}{Total\_number\_of\_docs\_in\_that\_month}$<br>

**Note 1:** If a document contains multiple keywords then include that document for computing the count and normalized count for all the keywords present in the document.<br>
**Note 2:** For a given `keyword` and `month`, if no document mentions that keyword in that month then return it's `doc_count` and `normalized_doc_count` as `0`.

**Note 3:** Include subwords for keyword count

**Input:**<br>
`input_df`
*  `uri: string`
*  `lang: string`
*  `crawl_month: string`
*  `content: string`

`keyword_df`
*  `keyword: string`

**Output:** Return a dataframe having the following schema:

*  `keyword: string`
*  `month: string`
*  `doc_count: long`
*  `normalized_doc_count: double`

```python
#######################################
###!@2.2 START ANSWER SET 2.2

### Q2.2 ###################################################

def question_2_2(input_df, keyword_df):

  ## start your edits here =================

  keyword_list = keyword_df.withColumn("keyword", F.lower("keyword")).select("keyword").distinct().rdd.flatMap(lambda x: x).collect()

  rows = []
  schema = StructType(
      [StructField("keyword", StringType(), True),
      StructField("month", StringType(), True),
      StructField("doc_count", LongType(), True),
      StructField("normalized_doc_count", DoubleType(), True)]
  )
  output_df = spark.createDataFrame(rows, schema)

  monthly_aggregated_df = input_df.groupBy("crawl_month").agg(
    F.count("*").alias("total_docs"),
    F.collect_list("content").alias("contents")
  )
  collected_data = monthly_aggregated_df.collect()
  monthly_data = {
      row["crawl_month"]: {
          "total_docs": row["total_docs"],
          "contents": row["contents"]
      } for row in collected_data
  }

  for month in monthly_data.keys():
    total_docs = monthly_data[month]["total_docs"]
    contents = monthly_data[month]["contents"]

    for keyword in keyword_list:
      doc_count = 0
      for content in contents:
        if keyword in content:
          doc_count += 1

      normalized_doc_count = 0 if total_docs == 0 else doc_count / total_docs
      rows.append((keyword, month, doc_count, normalized_doc_count))

  output_df = output_df.union(spark.createDataFrame(rows, schema))

  ## end your edits here =================

  return output_df

###!@2.2 END ANSWER SET 2.2
```

### Question 3: Popular, Trending and Emerging Topic Identification ***(15+15+15=45 points)***
----

In this question, you will build upon the normalized keyword statistics from earlier to distinguish between *popular, trending and emerging topics*. These help us understand not just what is being widely discussed, but also what is gaining momentum over time. A description of these trends are:
1. **Popular Topic:** It consistently generates a high volume of interest over a long period. Its defining characteristic is sustained engagement rather than sudden growth.
2. **Trending Topic:** A trending topic experiences a rapid and significant surge in attention, becoming a dominant subject of conversation for a short time. It is defined by its high rate of growth, not necessarily its longevity.
3. **Emerging Topics:** An emerging topic is a subject in the nascent stages of gaining traction, showing an accelerating rate of interest that suggests it may become a trend. It is characterized by its novelty and early-stage growth.

#### Question 3.1: Listing the Popular Topics **(15 points)**
For the given 4-month period `(April-2025 to July-2025)`, you have to identify the popular topics by performing the following tasks:
1. For each `keyword`, compute the average of the `normalized_doc_count` across the time period `(April-2025 to July-2025)` and store it as a column `avg_normalized_doc_count`.
2. Create four new columns, one for each quartile (`0-25`, `26-50`, `51-75`, `76-100`) and classify the `avg_normalized_doc_count` value into one of these four quartiles. You will return this as a "bit vector" row, i.e., for the quartile where the `avg_normalized_doc_count` value of keyword falls, assign the value `true`; for all other quartiles, assign value `false`. **Note:** The data type of quartiles column should be `bool`.

**Input:** You are given the dataframe from the output of *Q2.2* having schema:

*  `keyword: string`
*  `month: string`
*  `doc_count: long`
*  `normalized_doc_count: double`

**Output:** Return an output dataframe having schema:

*  `keyword: string`
*  `avg_normalized_doc_count: double`
*  `25_percentile: bool`
*  `50_percentile: bool`
*  `75_percentile: bool`
*  `100_percentile: bool`

```python
#######################################
###!@3.1 START ANSWER SET 3.1

### Q3.1 ###################################################

def question_3_1(input_df):

  ## start your edits here =================

  months_of_interest = ["2025-04", "2025-05", "2025-06", "2025-07"]

  output_df = input_df.filter(F.col("month").isin(months_of_interest)).groupBy("keyword").agg(F.avg(F.col("normalized_doc_count")).alias("avg_normalized_doc_count"))
  quantiles = output_df.approxQuantile("avg_normalized_doc_count", [0.25, 0.50, 0.75, 1.0], 0)

  output_df = output_df.withColumn("25_percentile", F.col("avg_normalized_doc_count") == 0)
  output_df = output_df.withColumn("50_percentile", (F.col("avg_normalized_doc_count") > 0) & (F.col("avg_normalized_doc_count") <= quantiles[1]))
  output_df = output_df.withColumn("75_percentile", (F.col("avg_normalized_doc_count") > quantiles[1]) & (F.col("avg_normalized_doc_count") <= quantiles[2]))
  output_df = output_df.withColumn("100_percentile", (F.col("avg_normalized_doc_count") > quantiles[2]) & (F.col("avg_normalized_doc_count") <= quantiles[3]))

  ## end your edits here =================

  return output_df

###!@3.1 END ANSWER SET 3.1
```

##Background

Large Language Models (LLMs) are powerful but have inherent limitations, including knowledge cutoffs (outdated information), a tendency to *hallucinate* facts, and a lack of source verifiability. Retrieval-Augmented Generation (RAG) is an architectural approach that mitigates these weaknesses by connecting the LLM to an external knowledge base. A typical RAG pipeline operates in three stages:<br>

* Retrieve: Relevant documents are fetched from a knowledge base in response to a query.
* Augment: The retrieved documents are inserted into the LLM's prompt, providing it with relevant context.
* Generate: The LLM generates an answer based on the augmented prompt.

Retrieving relevant data from a knowledge base involves the following components:
* Embedding Models: Converts text into numerical vectors that capture semantic meaning.
* Vector Databases: Store document chunks and their embeddings, enabling similarity-based retrieval.
* Similarity Search: Finds the most relevant documents by comparing vector representations.

In this question, you will focus on implementing the retrieval component of a RAG pipeline. Your task is to design and build a vector search mechanism using PySpark. #### Question 5.1: Prepare Vector Store **(10 points)**

Work with the given dataset `input_df` to prepare for a vector store using Spark Dataframes. For each of the web `content`, preprocess the text by splitting it into smaller *chunks* using a simple chunking logic implemented as follows. The content belonging to each `uri` will be splitted into chunks, such that the first chunk has the *first 256 characters*, the second chunk has the next *256 characters* and so on until the last chunk that has *at most* 256 characters. <br>

Webpages often contain large volumes of text, making it computationally expensive to generate embeddings for all content. To address this, we limit the chunks to at most 5 chunks that are evenly distributed across the content. The chunk selection logic is as follows:

* If a webpage has 5 or fewer chunks, include all of them.
* If a webpage has more than 5 chunks, select exactly 5 using the formula


$$
\text{index}_i = \left\lfloor \frac{i \times n}{5} \right\rfloor
$$

where:
- **i** in \{0,1,2,3,4\} is the position of the selected chunk.
- **n** is the total number of chunks in the document.

**Suggestion** use user defined functions, UDFs to implement the solution


**Input:**
The input to the question is a dataframe `input_df` which is the output of Questions `2.1`. The `input_df` has the following schema :

`input_df`

* `lang: string`

* `content: string`

**Output:**
Output a single dataframe that contains two columns `uri` and `chunks` where `chunks` contains the list of sampled chunks for that uri.

`output_df`
* `uri: string`
* `chunks: array<string>`


```python
#######################################
###!@5.1 START ANSWER SET 5.1

### Q5.1 ###################################################

def question_5_1(input_df):

  ## start your edits here =================

  chunk_length = 256
  max_chunks = 5
  output_df = input_df.withColumn("total_chunks", F.ceil(F.try_divide(F.length(F.col("content")), F.lit(chunk_length))))

  columns = []
  for i in range(max_chunks):
      columns.append(f"chunk_{i}")
      output_df = output_df.withColumn(f"chunk_{i}", F.substr(F.col("content"), F.try_add(F.try_multiply(F.floor(F.try_divide(F.try_multiply(F.lit(i), F.col("total_chunks")), F.lit(max_chunks))), F.lit(chunk_length)), F.lit(1)), F.lit(chunk_length)))

  output_df = output_df.withColumn("chunks", F.array_distinct(F.array(columns))).drop(*columns, "total_chunks", "lang", "content", "crawl_month")

  ## end your edits here =================

  return output_df

###!@5.1 END ANSWER SET 5.1
```

The *Cosine similarity* between two vectors $\mathbf{u}$ and $\mathbf{v}$ is defined as:<br><br>

$$
\operatorname{cosine\_similarity}(\mathbf{u}, \mathbf{v}) = \frac{\mathbf{u} \cdot \mathbf{v}}{\Vert\mathbf{u}\Vert \, \Vert\mathbf{v}\Vert}$$

- $\mathbf{u} \cdot \mathbf{v}$ is the **dot product** of the two vectors.  
- $\|\mathbf{u}\|$ and $\|\mathbf{v}\|$ are their **Euclidean norms** (vector lengths).  
- The value is always between **-1 and 1**:
- **1** means same direction (high similarity)  
- **0** means orthogonal (no similarity)  
- **-1** means opposite direction (opposite meaning)
