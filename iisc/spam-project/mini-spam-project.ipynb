```python
from pyspark.sql import SparkSession

sc = SparkSession.builder.appName("SpamProject").master("spark://localhost:7077").getOrCreate()
print(sc)
```

```python
from pyspark.sql import functions as F
from pyspark.sql.types import IntegerType, StringType, StructField, StructType

```

```python
import os
os.getcwd()
```

```python
import pandas as pd
df = pd.read_csv(f'{os.getcwd()}/spam-project.txt')
```

```python
csvSchema = StructType([
    StructField("label", StringType(), nullable=False),
    StructField("message", StringType(), nullable=False)
])

# Create the spark DF using the pandas DF
rawDF = sc.createDataFrame(df, schema=csvSchema)
rawDF.show(10)
```
